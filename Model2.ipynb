```python
dummies=pd.get_dummies(df.State)
dummies=dummies.astype(int)
dummies.columns
```

```python
df2=pd.concat([df,dummies],axis=1)
df2.head()
```

```python
dummies=pd.get_dummies(df["Commodity"])
dummies=dummies.astype(int)
dummies.columns
```

```python
df3=pd.concat([df2,dummies],axis=1)
df3.head()
```

```python
df4=df3.drop(["Commodity","State"],axis=1)
df4.head()
```

```python
df4[["Month","Year"]] = df4['Date'].str.split('-',n=1,expand=True)
df4.head()
```

```python
def split_Date(val):
    mon = {'JAN': 1, 'FEB': 2, 'MAR': 3, 'APR': 4, 'MAY': 5, 'JUN': 6, 'JUL': 7, 'AUG': 8, 'SEP': 9, 'OCT': 10, 'NOV': 11, 'DEC': 12}
    return mon[val.upper()]
```

```python
df4["Month"] = df4["Month"].apply(split_Date)
```

```python
df4.head()
```

```python
df5=df4.drop(["Date"],axis=1)
df5.head()
```

```python
df5.reset_index(drop=True, inplace=True)
df5.head()
```

```python
X=df5.drop(["Retail Price"],axis=1)
X.head()
```

```python
Y=df5["Retail Price"]
Y.head()
```

```python
X_train , X_test,Y_train , Y_test = train_test_split(X,Y,test_size=0.2)
```

```python
model=linear_model.LinearRegression()
```

```python
model.fit(X_train,Y_train)
```

```python
model.score(X_test,Y_test)
```

```python
def predict_price(State,Commodities,Month,Year):    
    loc_index = np.where(X.columns==State)[0][0]
    loc_index2=np.where(X.columns==Commodities)[0][0]

    x = np.zeros(len(X.columns))
    if loc_index >= 0:
        x[loc_index] = 1
    if loc_index2 >= 0:
        x[loc_index2] = 1
    x[76] = Month
    x[77] = Year
    if loc_index >= 0:
        x[loc_index] = 1

    return model.predict([x])[0]
```
